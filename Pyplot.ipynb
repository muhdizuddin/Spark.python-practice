```python
def f(t):
  return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.02)
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(t1, f(t1),'bo')

plt.subplot(2,1,2)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')# using fuctions and model

plt.ylabel("Y label");
```

```python
x = np.linspace(-4,4)
y1, y2 = x**2, x**3
```

```python
#object orianted interface
fig, ax = plt.subplots()
ax.plot(x, y1, 'red')
ax.plot(x, y2, 'blue')
ax.set_title('Quadratic and cubic powers')
ax.set_xlabel('Xlabel')
ax.set_ylabel('Ylabel');
```

```python
#pyplot
plt.plot(x, y1, 'red', x, y2, 'blue')
plt.title('Quadratic and cubic powers')
plt.xlabel('Xlabel')
plt.ylabel('Ylabel');
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0,0].set_title('Uleft')
axes[0,1].set_title('Uright')
axes[1,0].set_title('Lleft')
axes[1,1].set_title('Lright')
for ax in axes.flat:
  #remove all xticks and y ticks
  ax.set(xticks=[],yticks=[])

fig.tight_layout();
```

```python
#creating axes/subplots in one fucntions
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(7,4.5))
axes[0,0].set_title('Linear')
axes[0,0].plot(x,x)
axes[0,1].set_title('Quadratic')
axes[0,1].plot(x,x**2)
axes[1,0].set_title('Cubic')
axes[1,0].plot(x,x**3)
axes[1,1].set_title('4th Power')
axes[1,1].plot(x,x**4)
fig.tight_layout();
```

```python
fig, axes = plt.subplots(nrows=2,ncols=5,figsize=(14,4.5))
for i, ax in enumerate(axes.flatten()):
  ax.set_title("x to the {:d}".format(i+1))
  ax.plot(x,x**(i+1))

fig.tight_layout();
```

```python
x = np.linspace(-np.pi, np.pi, 200)
sine,cosine = np.sin(x), np.cos(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x,sine,x,cosine);
```

```python
#elobrating the chart 
fig, ax =plt.subplots()
ax.plot(x, sine, color = 'red', label='sine')
ax.plot(x, cosine, color = 'blue', label='cosine')

#axes limit
ax.set_xlim(-3,5,3.5)
ax.set_ylim(-1.2,1,2)

#changing marks accroding to specific value
ax.set_xticks([-np.pi, -np.pi/2, 0 ,np.pi/2,np.pi])
ax.set_yticks([-1,0,1])

#appears label as mathematical notations
ax.set_xticklabels([r'$-\pi$',r'$-\pi/2$',r'$0$',r'$+\pi/2$',r'$+\pi$'],size=17)
ax.set_yticklabels(['-1','0','+1'],size=17)

#ploting clue
ax.legend(loc='upper left')

#item annotate
ax.text(-0.25,0,'(0,0)')
ax.text(np.pi-0.2,0.05,r'$(\pi,0)$',size=15)

#arrow annotate
ax.annotate('Origin',
            xy=(0,0),
            xytext=(1,-0.7),
            arrowprops=dict(facecolor='blue'));

#vertical and horizontol line
ax.axhline(0, color='black',alpha=0.9)#hor
ax.axvline(0, color='black',alpha=0.9)#ver
ax.grid();#grid line
```
